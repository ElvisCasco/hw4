# a. data_loader.py

```python
import pandas as pd

def data_loader(file_name: str) -> pd.DataFrame:
    data = pd.read_csv(file_name)
    print(f"Data loaded successfully. Shape: {data.shape}")
    return data
```

```python
FILE= 'sample_diabetes_mellitus_data.csv'
```

```python
data_loader(FILE)
```

# b. data_split.py

```python
from sklearn.model_selection import train_test_split
def data_split(file_name):
    data=data_loader(file_name)
    train, test = train_test_split(data, test_size=0.2, random_state=42)
    return train,test
```

```python
train,test=data_split(FILE)
```

```python
train
```

```python
test
```

# c. data_remove_nans.py

```python
def data_remove_nans(file_name):
    train,test=data_split(file_name)
    train = train.dropna(subset=['age', 'gender', 'ethnicity'])
    test = test.dropna(subset=['age', 'gender', 'ethnicity'])
    return train,test
```

```python
train,test=data_remove_nans(FILE)
```

```python
train
```

```python
test
```

# d. data_fill_nans.py

```python
def data_fill_nans(file_name):
    train,test=data_remove_nans(file_name)
    train['height'] = train['height'].fillna(train['height'].mean())
    train['weight'] = train['weight'].fillna(train['weight'].mean())
    test['height'] = test['height'].fillna(test['height'].mean())
    test['weight'] = test['weight'].fillna(test['weight'].mean())
    return train,test
```

```python
train,test=data_fill_nans(FILE)
```

```python
train.isna().sum()
```

```python
test.isna().sum()
```

# e. data_encoding.py

```python
#e.Generate dummies for ethnicity column (One hot encoding).
def data_encoding(file_name):
    train,test=data_fill_nans(file_name)
    train = pd.get_dummies(train, columns=['ethnicity'], drop_first=True)
    test = pd.get_dummies(test, columns=['ethnicity'], drop_first=True)
    return train,test
```

```python
train,test=data_encoding(FILE)
```

```python
train
```

```python
test
```

# f. data_binary.py

```python
#f.Create a binary variable for gender M/F.
def data_binary(file_name):
    train,test= data_encoding(file_name)
    train['gender_M'] = (train['gender'] == 'M').astype(int)
    train['gender_F'] = (train['gender'] == 'F').astype(int)
    test['gender_M'] = (test['gender'] == 'M').astype(int)
    test['gender_F'] = (test['gender'] == 'F').astype(int)
    return train, test
```

```python
train, test = data_binary(FILE)
```

# g. & h: data_predict.py

## g. Train a model (logistic)

```python
from sklearn.linear_model import LogisticRegression
def data_logistic_model(file_name):
     # Get the logistic regression model using training data
     train,test=data_binary(file_name)
     features = [
         'age', 'height', 'weight', 'aids', 'cirrhosis',
          'hepatic_failure','immunosuppression', 'leukemia',
          'lymphoma', 'solid_tumor_with_metastasis']
     # Get features and target variables
     X_train = train[features]
     y_train = train['diabetes_mellitus']
     X_test = test[features]
     y_test = test['diabetes_mellitus']
     # Fit the model with training data
     model = LogisticRegression(max_iter=1000)
     model.fit(X_train, y_train)
     return X_train,y_train,X_test,y_test,model
```

```python
X_train,y_train,X_test,y_test,model = data_logistic_model(FILE)
```

## h. Predict for train and test using the logistic model

```python
def data_predict(file_name):
     train,test=data_binary(file_name)
     X_train,y_train,X_test,y_test,model=data_logistic_model(file_name)
     # Predict class labels and probabilities
     train['predictions'] = model.predict(X_train)
     train['predictions_prob'] = model.predict_proba(X_train)[:, 1]
     test['predictions'] = model.predict(X_test)
     test['predictions_prob'] = model.predict_proba(X_test)[:, 1]
     return train,test
```

```python
train,test=data_predict(FILE)
```

# i. Compute the train and test roc_auc metric using roc_auc_score from sklearn.

```python
from sklearn.metrics import roc_auc_score, accuracy_score

def pred_auc_score(file_name):
     train,test=data_predict(file_name)
     # Evaluate performance
     y_train = train['diabetes_mellitus']
     y_test = test['diabetes_mellitus']
     train_auc = roc_auc_score(y_train, train['predictions_prob'])
     #train_accuracy = accuracy_score(y_train, train['predictions'])
     test_auc = roc_auc_score(y_test, test['predictions_prob'])
     #test_accuracy = accuracy_score(y_test, train['predictions'])
     # Print results
     print("Results from forecasting diabetes_mellitus using logistic model:\n")
     print(f"Train AUC: {train_auc:.4f}\n")
     #print(f"Train Accuracy: {train_accuracy:.4f}")
     print(f"Test AUC: {test_auc:.4f}\n")
     #print(f"\nTest Accuracy: {test_accuracy:.4f}")
```
